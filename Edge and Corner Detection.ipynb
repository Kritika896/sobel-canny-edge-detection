```python
# Edge Detection using U-Net in Keras with Dice Loss
import numpy as np
import matplotlib.pyplot as plt
import tensorflow as tf
from tensorflow.keras.layers import Input, Conv2D, MaxPooling2D, UpSampling2D, concatenate
from tensorflow.keras.models import Model
from sklearn.model_selection import train_test_split
import cv2
import os
from tqdm import tqdm

# === Load and preprocess images ===
def load_images_from_folder(folder, size=(128, 128)):
    images = []
    for filename in tqdm(os.listdir(folder)):
        img = cv2.imread(os.path.join(folder, filename), cv2.IMREAD_GRAYSCALE)
        if img is not None:
            img = cv2.resize(img, size)
            img = img / 255.0
            images.append(img)
    return np.array(images)

image_folder = r"C:\Users\kriti\Downloads\archive (5)\train_zip\train"  # <- Set your image folder path
images = load_images_from_folder(image_folder)
images = np.expand_dims(images, axis=-1)

# === Generate edge maps using Sobel ===
def generate_sobel_edges(images):
    edges = []
    for img in images:
        img = (img * 255).astype(np.uint8)
        sobelx = cv2.Sobel(img, cv2.CV_64F, 1, 0, ksize=3)
        sobely = cv2.Sobel(img, cv2.CV_64F, 0, 1, ksize=3)
        edge = np.sqrt(sobelx**2 + sobely**2)
        edge = cv2.normalize(edge, None, 0, 1, cv2.NORM_MINMAX)
        edges.append(edge)
    return np.expand_dims(np.array(edges), axis=-1)

edges = generate_sobel_edges(images)

# === Train-test split ===
X_train, X_test, y_train, y_test = train_test_split(images, edges, test_size=0.2, random_state=42)

# === Define Dice Loss ===
def dice_loss(y_true, y_pred):
    smooth = 1.
    y_true_f = tf.keras.backend.flatten(y_true)
    y_pred_f = tf.keras.backend.flatten(y_pred)
    intersection = tf.reduce_sum(y_true_f * y_pred_f)
    return 1 - ((2. * intersection + smooth) /
                (tf.reduce_sum(y_true_f) + tf.reduce_sum(y_pred_f) + smooth))

# === Define U-Net Model ===
def unet_model(input_shape):
    inputs = Input(input_shape)

    # Encoder
    c1 = Conv2D(16, (3, 3), activation='relu', padding='same')(inputs)
    p1 = MaxPooling2D((2, 2))(c1)

    c2 = Conv2D(32, (3, 3), activation='relu', padding='same')(p1)
    p2 = MaxPooling2D((2, 2))(c2)

    # Bottleneck
    b = Conv2D(64, (3, 3), activation='relu', padding='same')(p2)

    # Decoder
    u2 = UpSampling2D((2, 2))(b)
    concat2 = concatenate([u2, c2])
    c3 = Conv2D(32, (3, 3), activation='relu', padding='same')(concat2)

    u1 = UpSampling2D((2, 2))(c3)
    concat1 = concatenate([u1, c1])
    c4 = Conv2D(16, (3, 3), activation='relu', padding='same')(concat1)

    outputs = Conv2D(1, (1, 1), activation='sigmoid')(c4)

    return Model(inputs, outputs)

# === Compile the model ===
model = unet_model(X_train.shape[1:])
model.compile(optimizer='adam', loss=dice_loss, metrics=['accuracy'])
model.summary()

# === Train the model ===
history = model.fit(X_train, y_train, epochs=20, batch_size=8, validation_split=0.1)

# === Predict on test data ===
preds = model.predict(X_test)

# === Visualize results ===
num = 5  # Number of test images to display
plt.figure(figsize=(15, 5))
for i in range(num):
    plt.subplot(3, num, i+1)
    plt.imshow(X_test[i].squeeze(), cmap='gray')
    plt.title("Original")
    plt.axis('off')

    plt.subplot(3, num, i+1+num)
    plt.imshow(y_test[i].squeeze(), cmap='gray')
    plt.title("Edge GT")
    plt.axis('off')

    plt.subplot(3, num, i+1+2*num)
    plt.imshow(preds[i].squeeze() > 0.5, cmap='gray')  # Thresholded prediction
    plt.title("Predicted")
    plt.axis('off')

plt.tight_layout()
plt.show()
```
